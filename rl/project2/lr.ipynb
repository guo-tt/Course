```python
import gym
import numpy as np
import pandas as pd
import random

import pickle
from matplotlib import pyplot as plt

from network import DQN
```

```python
def plot_mean(df, name, x_label, y_label, y_limit):
    plt.rcParams.update({'font.size': 20})
    plt.figure(figsize=(20, 12)) 
    plt.close()
    plt.figure()
    
    plot = df.plot(linewidth=3.0, figsize=(20, 12))
    plot.set_xlabel(x_label) 
    plot.set_ylabel(y_label)
    plt.ylim(y_limit)
    fig = plot.get_figure()
    fig.savefig('fig/'+ name)
```

```python
env = gym.make('LunarLander-v2')

# set seeds
env.seed(6)
np.random.seed(6)

# setting up params
values = [0.0001, 0.001, 0.01]
epsilon = 1.0
decay = 0.995
gamma = 0.99
training_episodes = 1000 

r_lrs = []

save_dir = "models/"

for lr in values:
    model = DQN(env, lr, gamma, epsilon, decay)
    print("Training model for LR: {}".format(lr))
    model.train(training_episodes, False)
    r_lrs.append(model.r_list)
    
    model.save(save_dir + "lr_model_" + str(lr) + ".h5") 
    
```

```python
pickle.dump(r_lrs, open("models/" + "rewards_lr.p", "wb"))
```

```python
rewards_lrs = pickle.load(open("models/" + "rewards_lr.p", "rb"))
```

```python
lr_rewards_pd = pd.DataFrame(index=pd.Series(range(1, training_episodes + 1)))

for i in range(len(values)):
    col_name = "lr="+ str(values[i])
    lr_rewards_pd[col_name] = rewards_lrs[i]
plot_mean(lr_rewards_pd, "lr", "episodes", "reward", (-6000, 500)) 
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n 

ix = 0 
for i in range(len(values)):
    col_name = "lr="+ str(values[i])
    lr_rewards_pd[col_name] = rewards_lrs[i]
    print("lr: ", values[i])
    print("mean: ",np.mean(rewards_lrs[i]))
    print("%: ", sum(i > 200 for i in rewards_lrs[i])/len(rewards_lrs[i]))
    temp = moving_average(rewards_lrs[i],100)
    try:
        print("begin to converge at: ", next(x for x, val in enumerate(temp) if val > 200) )
    except:
        print("no > 200") 
```
