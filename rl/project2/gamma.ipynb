```python
import gym
import numpy as np
import pandas as pd
import random

import pickle
from matplotlib import pyplot as plt

from network import DQN
```

```python
def plot_mean(df, name, x_label, y_label, y_limit):
    plt.rcParams.update({'font.size': 20})
    plt.figure(figsize=(20, 12)) 
    plt.close()
    plt.figure()
    
    plot = df.plot(linewidth=3.0, figsize=(20, 12))
    plot.set_xlabel(x_label) 
    plot.set_ylabel(y_label)
    plt.ylim(y_limit)
    fig = plot.get_figure()
    fig.savefig('fig/'+ name)
```

```python
print('Running Experiment for gamma...')
env = gym.make('LunarLander-v2')

# set seeds
env.seed(6)
np.random.seed(6)

# setting up params
lr = 0.001
epsilon = 1.0
decay = 0.995
gamma_list = [0.999, 0.99, 0.90]
training_episodes = 1000

r_gammas = []
save_dir = "models/"

for gamma in gamma_list:
    # save_dir = "hp_gamma_"+ str(gamma_value) + "_"
    model = DQN(env, lr, gamma, epsilon, decay)
    model.train(training_episodes, False)

    model.save(save_dir + "gamma_model_" + str(gamma) + ".h5")
    r_gammas.append(model.r_list)
    
pickle.dump(r_gammas, open(save_dir + "rewards_for_gammas.p", "wb"))
```

```python
rewards_gammas = pickle.load(open(save_dir + "rewards_for_gammas.p", "rb"))

gamma_rewards_pd = pd.DataFrame(index=pd.Series(range(1, training_episodes + 1)))

for i in range(len(gamma_list)):
    col_name = r"$\gamma$=" + str(gamma_list[i])
    gamma_rewards_pd[col_name] = rewards_gammas[i]
    
plot_mean(gamma_rewards_pd, "gamma", "episodes", "reward", (-7000, 350))  
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n 

ix = 0 
for i in range(len(gamma_list)):
    col_name = "lr="+ str(gamma_list[i])
    gamma_rewards_pd[col_name] = rewards_gammas[i]
    print("lr: ", gamma_list[i])
    print("mean: ",np.mean(rewards_gammas[i]))
    print("%: ", sum(i > 200 for i in rewards_gammas[i])/len(rewards_gammas[i]))
    temp = moving_average(rewards_gammas[i],100)
    try:
        print("begin to converge at: ", next(x for x, val in enumerate(temp) if val > 200) )
    except:
        print("no > 200") 
```
