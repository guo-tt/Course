```python
import gym
import numpy as np
import pandas as pd
import random

import pickle
from matplotlib import pyplot as plt

from network import DQN
```

```python
def plot_mean(df, name, x_label, y_label, y_limit):
    plt.rcParams.update({'font.size': 20})
    plt.figure(figsize=(20, 12)) 
    plt.close()
    plt.figure()
    
    plot = df.plot(linewidth=3.0, figsize=(20, 12))
    plot.set_xlabel(x_label) 
    plot.set_ylabel(y_label)
    plt.ylim(y_limit)
    fig = plot.get_figure()
    fig.savefig('fig/'+ name)
```

```python
print('epsilon decay...')
env = gym.make('LunarLander-v2')

# set seeds
env.seed(2134)
np.random.seed(2134) 

# setting up params
lr = 0.001
epsilon = 1.0
values = [0.999, 0.995, 0.990]
gamma = 0.99
training_episodes = 1000

rewards_ed = []
save_dir = "models/"

for ed in values:
    model = DQN(env, lr, gamma, epsilon, ed)
    print("Training model for ED: {}".format(ed))
    model.train(training_episodes, False)

    model.save(save_dir + "ed_model_" + str(ed) + ".h5")
    rewards_ed.append(model.r_list) 
    
pickle.dump(rewards_ed, open("models/"+"rewards_for_ed.p", "wb"))    
```

```python
rewards_ed = pickle.load(open("models/"+"rewards_for_ed.p", "rb"))

ed_rewards_pd = pd.DataFrame(index=pd.Series(range(1, training_episodes+1)))
for i in range(len(values)):
    col_name = r"$\epsilon$ decay = "+ str(ed_values[i])
    ed_rewards_pd[col_name] = rewards_ed[i]
    
plot_mean(ed_rewards_pd, "ed", "episodes", "reward", (-900, 350)) 
```
