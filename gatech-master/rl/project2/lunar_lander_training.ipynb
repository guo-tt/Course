```python
from network import DQN

import gym
import numpy as np
import pandas as pd
from collections import deque
import random

import pickle
from matplotlib import pyplot as plt
```

```python
def plot_mean(df, name, x, y):
    plt.rcParams.update({'font.size': 20})
    df['mean'] = df[df.columns[0]].rolling(100).mean()
    plt.figure(figsize=(20, 12)) 
    plt.close()
    plt.figure()
    
    plot = df.plot(linewidth=3.0, figsize=(20, 12))
    plot.set_xlabel(x) 
    plot.set_ylabel(y)
    fig = plot.get_figure()
    plt.legend().set_visible(False) 
    fig.savefig('fig/'+ name) 
```

```python
env = gym.make('LunarLander-v2')

# set seeds
env.seed(6) 
np.random.seed(6)  

# setting up params
lr = 0.001
epsilon = 1.0
decay = 0.995
gamma = 0.99 
episodes = 500 

model = DQN(env, lr, gamma, epsilon, decay)
model.train(episodes, True)
```

```python
# Save Everything
dir = "models/"
model.save(dir + "model_softmax.h5")
```

```python
# Save Rewards list
pickle.dump(model.r_list, open(dir + "train_rewards_softmax.p", "wb"))
r_list = pickle.load(open(dir + "train_rewards_softmax.p", "rb"))

# plot reward in graph
reward_df = pd.DataFrame(r_list)
plot_mean(reward_df, "train_softmax", "episode","reward")
```
