```python
import gym
import numpy as np
import pandas as pd
import random

import pickle
from matplotlib import pyplot as plt

from network import DQN
```

```python
def plot_mean(df, name, x_label, y_label, y_limit):
    plt.rcParams.update({'font.size': 20})
    plt.figure(figsize=(20, 12)) 
    plt.close()
    plt.figure()
    
    plot = df.plot(linewidth=3.0, figsize=(20, 12))
    plot.set_xlabel(x_label) 
    plot.set_ylabel(y_label)
    plt.ylim(y_limit)
    fig = plot.get_figure()
    fig.savefig('fig/'+ name)
```

```python
env = gym.make('LunarLander-v2')

# set seeds
env.seed(6)
np.random.seed(6) 

# setting up params
lr = 0.001
epsilon = 1.0
values = [0.999, 0.995, 0.990]
gamma = 0.99
training_episodes = 1000

rewards_ed = []
save_dir = "models/"

for ed in values:
    model = DQN(env, lr, gamma, epsilon, ed)
    model.train(training_episodes, False)

    model.save(save_dir + "ed_model_" + str(ed) + ".h5")
    rewards_ed.append(model.r_list) 
    
pickle.dump(rewards_ed, open("models/"+"rewards_ed.p", "wb"))    
```

```python
rewards_ed = pickle.load(open("models/"+"rewards_ed.p", "rb"))

ed_rewards_pd = pd.DataFrame(index=pd.Series(range(1, training_episodes+1)))
for i in range(len(values)):
    col_name = r"$\epsilon$ decay = "+ str(values[i])
    ed_rewards_pd[col_name] = rewards_ed[i]
    
plot_mean(ed_rewards_pd, "ed", "episodes", "reward", (-900, 350)) 
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n 

ix = 0 
for i in range(len(values)):
    col_name = "lr="+ str(values[i])
    ed_rewards_pd[col_name] = rewards_ed[i]
    print("lr: ", values[i])
    print("mean: ",np.mean(rewards_ed[i]))
    print("%: ", sum(i > 200 for i in rewards_ed[i])/len(rewards_ed[i]))
    temp = moving_average(rewards_ed[i],100)
    try:
        print("begin to converge at: ", next(x for x, val in enumerate(temp) if val > 200) )
    except:
        print("no > 200") 
```
